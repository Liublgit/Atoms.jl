```julia
using Devectorize

function lj(r)
    t = 1.0 ./ r
    t = t .* t
    t = t .* t .* t
    sum(t.*t - 2.0*t)
end

function lj4(r)
    s = 0.0
    @simd for n = 1:length(r) @inbounds begin
            r[n] = 1.0 / (r[n]*r[n])
            r[n] = r[n]*r[n]*r[n]
            r[n] = (r[n]-2.0) * r[n]
            s = s + r[n]
        end
    end
    return s
end

function lj3(r)
    @devec r[:] = 1 ./ r[:]
    @devec r[:] = r[:] .* r[:]
    @devec r[:] = r[:] .* r[:] .* r[:]
    sum(r.*r - 2.0 * r)
end

function lj2(r)
    sum(r.^(-12) - 2.0*r.^(-6))
end
```
